## Sklearn.metrics.pairwise_distances can be used to compute distance between all pairs of users.  pairwise_distances()  takes  a  metric  parameter  for  what  distance  measure  to  use. Use cosine similarity for finding similarity among users. Use the following packages.

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
distance_matrix = pairwise_distances(df, metric='cosine')
```

## Find the 5 most similar user for user with user Id 25.

```python
(distance_matrix[24]).argsort()[1:6]+1
```

```python
moviesub=set(df_ratings.loc[df_ratings.userId==1, 'movieId']).intersection(set(df_ratings.loc[df_ratings.userId==338, 'movieId']))
for i in moviesub:
    print(df_movies.loc[df_movies.movieId==i, 'title'].values[0])
```

## Use the “movies” dataset to find out the names of movies, user 1 and user 338 have watched in common and how they have rated each one of them.

```python
for i in moviesub:
    print('Movie ID ',i,": ", df_movies.loc[df_movies.movieId==i, 'title'].values[0])
    print('Rating by User 1: ', df_ratings.loc[((df_ratings.userId==1) & (df_ratings.movieId==i)), 'rating'].values[0])
    print('Rating by User 338: ', df_ratings.loc[((df_ratings.userId==338) & (df_ratings.movieId==i)), 'rating'].values[0])
```

## Use the movies dataset to find out the common movie names between user 2 and user 338 with least rating of 4.0

```python
 moviesub=set(df_ratings.loc[((df_ratings.userId==2) & (df_ratings.rating>=4.0)), 'movieId']).intersection(set(df_ratings.loc[((df_ratings.userId==338) & (df_ratings.rating>=4.0)), 'movieId']))
for i in moviesub:
    print(df_movies.loc[df_movies.movieId==i, 'title'].values[0])
```
